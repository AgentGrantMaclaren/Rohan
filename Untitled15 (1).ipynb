```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[9,2,3,4]
y=[19,10,7,30]
plt.plot(x,y)
plt.show()
```

```python
x=[9,2,3,4]
y=[19,10,7,30]
plt.bar(x,y)
plt.show()
```

```python
sns.set(style="white")
rs=np.random.RandomState(10)
d=rs.normal(size=100)
sns.displot(d,kde=True,color="r")
```

```python
sns.set(style="dark")
fmri=sns.load_dataset("fmri")
sns.lineplot(x="timepoint",y="signal",hue="region",style="event",data=fmri)
```

```python
sns.set(style="ticks")
df=sns.load_dataset("anscombe")
sns.lmplot(x="x",y="y",data=df)
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
area=[600,800,1000,1200,1500]
price=[15,18,25,35,40]
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("area vs price")
plt.xlabel("Area in sq ft")
plt.ylabel("price in lakh")
plt.show()
```

```python
x=[100,200,300,400,500,600,700,800]
y=[110,200,900,666,780,90,60,999]
plt.figure(figsize=(7,7))
plt.plot(x,y)
plt.title("area vs price")
plt.xlabel("Area in sq ft")
plt.ylabel("price in lakh")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("area vs price")
plt.xlabel("Area in sq ft")
plt.ylabel("price in lakh")
plt.xlim((400,2000))
plt.ylim((10,50))
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("area vs price")
plt.xlabel("Area in sq ft")
plt.ylabel("price in lakh")
plt.xticks(rotation=45)
plt.yticks(rotation=90)
plt.show()
```

```python
x=np.arange(1,20)
y1=np.sqrt(x)
y2=np.log(x)
plt.figure(figsize=(7,7))
plt.plot(x,y1,color="blue")
plt.plot(x,y2,color="red")
plt.legend(labels=["square root","log"])
plt.show()
```
