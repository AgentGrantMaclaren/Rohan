```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("iris")
def graph():
    sns.lineplot(x="sepal_length",y="sepal_width",data=data)

plt.subplot(121)
graph()
plt.subplot(122)
graph()
plt.show()
```

```python
data1=sns.load_dataset("iris")
sns.boxplot(x="species",y="sepal_width",data=data1)
plt.show()
```

```python
data1=sns.load_dataset("iris")
sns.violinplot(x="species",y="sepal_width",data=data1)
plt.show()

```
